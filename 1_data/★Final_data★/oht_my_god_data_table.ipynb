```
# 데이터 프레임 합병
df = pd.merge(arranged,category_table)
df
```

```
# texture_1 ~ style_2 총 4개의 열에서 결측치를 가지고 있는 행 삭제
df_final = df.dropna(thresh=4, axis=0) # 비결측치를 4개 이상 가지는 행을 남긴다 (image_name + category_label + category_name + (texture_1 ~ style_2 중 하나 ) = 4 )
df_final
```

```
df_final_0 = df_final.fillna(0)                             # NaN값에 일단 0으로 채움.
df_final_0
```

```
t = df_final_0[ (df_final_0['texture_1'] == 0)                  #  174873에서   111052개로 줄어듬. / 'texture1', 'texture2' 값이 0인 것들은 지움.
                & (df_final_0['texture_2'] == 0) ].index
df_final_0_t = df_final_0.drop(t)
df_final_0_t
```

```
s = df_final_0_t[ (df_final_0_t['style_1'] == 0)                      #  111052에서   41143개로 줄어듬. / 'style_1', 'style_2' 값이 0인 것들은 지움.
                & (df_final_0_t['style_2'] == 0) ].index
df_final_0_s = df_final_0_t.drop(s)
df_final_0_s
```
